```python
Cat_V = sp.select_dtypes('object').columns
for sp_col in Cat_V:
  sp[Cat_V ] = sp[Cat_V].astype('category') #Modificación tipos de datos.
```

```python
int_Var = sp.select_dtypes('int64').columns
```

```python
for sp_col in int_Var:
  sp[sp_col] = sp[sp_col].astype('int16') #disminuir memoria en uso
```

```python
float_64_Var = sp.select_dtypes('float64').columns
```

```python
for sp_col in float_64_Var:
  sp[sp_col] = sp[sp_col].astype('float16') #disminuir memoria en uso
```

```python
sp['ESTU_DEPTO_RESIDE'].fillna(sp['ESTU_DEPTO_RESIDE'].mode()[0], inplace=True)
sp['ESTU_MCPIO_RESIDE'].fillna(sp['ESTU_MCPIO_RESIDE'].mode()[0], inplace=True)
sp['FAMI_ESTRATOVIVIENDA'].fillna(sp['FAMI_ESTRATOVIVIENDA'].mode()[0], inplace=True)
sp['FAMI_EDUCACIONMADRE'].fillna(sp['FAMI_EDUCACIONMADRE'].mode()[0], inplace=True)
sp['FAMI_EDUCACIONPADRE'].fillna(sp['FAMI_EDUCACIONPADRE'].mode()[0], inplace=True)
sp['FAMI_TRABAJOLABORPADRE'].fillna(sp['FAMI_TRABAJOLABORPADRE'].mode()[0], inplace=True)
sp['FAMI_TRABAJOLABORMADRE'].fillna(sp['FAMI_TRABAJOLABORMADRE'].mode()[0], inplace=True)
sp['FAMI_TIENEINTERNET'].fillna(sp['FAMI_TIENEINTERNET'].mode()[0], inplace=True)
sp['FAMI_TIENECOMPUTADOR'].fillna(sp['FAMI_TIENECOMPUTADOR'].mode()[0], inplace=True)
sp['FAMI_SITUACIONECONOMICA'].fillna(sp['FAMI_SITUACIONECONOMICA'].mode()[0], inplace=True)
sp['ESTU_DEDICACIONLECTURADIARIA'].fillna(sp['ESTU_DEDICACIONLECTURADIARIA'].mode()[0], inplace=True)
sp['ESTU_DEDICACIONINTERNET'].fillna(sp['ESTU_DEDICACIONINTERNET'].mode()[0], inplace=True)
sp['PUNT_INGLES'].fillna(sp['PUNT_INGLES'].mode()[0], inplace=True)
sp['ESTU_NSE_INDIVIDUAL'].fillna(sp['ESTU_NSE_INDIVIDUAL'].mode()[0], inplace=True)
sp['ESTU_HORASSEMANATRABAJA'].fillna(sp['ESTU_HORASSEMANATRABAJA'].mode()[0], inplace=True)
```
